Take two integer numbers and return their product. if product is greater than 1000 then return their sum if product is less than 1000 product value only printed

```python
def fun(n1,n2):
  product=n1*n2
  if product>1000:
    print(n1+n2)
  else:
    print(n1*n2)
result=fun(20,10)
print(result)
```

2.given a range of first 10 numbers ,iterate from start number to the end number ,and each iteeration print the sum of current number and previous number

```python
previous=0
for current in range(1,11):
  sum=current+previous
  print(f"current number:{current},previous number:{previous},sum:{sum}")
  previous=current
```

given a string and an integer number n,remove character from a string starting from zero up to n and return a new string

```python
def fun(inputstring,n):
  if n<len(inputstring):
    resultstring=inputstring[n:]
    print(resultstring)
  else:
    print("n should be less than the length of string")
inputstring="souvikdas"
n=3
result=fun(inputstring,n)
print(result)
```

4.given list of numbers,return true if first and last numberof a list is same

```python
def fun(list):
  if len(list)>=2 and list[0]==list[-1]:
    print("true")
  else:
    print("false")
list=[1,2,3,6,7,1]
result=fun(list)
print(result)
```

5.given a two python list. iterate both lists simultaneously such that list1 should display item in original order and list2 reverse order

```python
list1 = [1, 2, 3, 4]
list2 = [5, 6, 7, 8]

for item1, item2 in zip(list1, reversed(list2)):
    print(f"List1: {item1}, List2 (Reversed): {item2}")
```
